# STORE ANALYSIS

```python
#Size Bucket vs Total Sales
query = """
    SELECT
        CASE
            WHEN square_meters < 250 THEN '<250'
            WHEN square_meters BETWEEN 250 AND 500 THEN '250 to 500'
            WHEN square_meters BETWEEN 501 AND 750 THEN '500 to 750'
            WHEN square_meters BETWEEN 751 AND 1000 THEN '750 to 1000'
            WHEN square_meters BETWEEN 1001 AND 1250 THEN '1000 to 1250'
            WHEN square_meters BETWEEN 1251 AND 1500 THEN '1250 to 1500'
            WHEN square_meters BETWEEN 1501 AND 1750 THEN '1500 to 1750'
            WHEN square_meters BETWEEN 1751 AND 2000 THEN '1750 to 2000'
            WHEN square_meters > 2000 THEN '> 2000'
        END AS size_bucket,
        ROUND(SUM(unit_price_usd * quantity),2) AS total_sales
    FROM overall
    GROUP BY size_bucket
    ORDER BY
        CASE
            WHEN size_bucket = '<250' THEN 1
            WHEN size_bucket = '250 to 500' THEN 2
            WHEN size_bucket = '500 to 750' THEN 3
            WHEN size_bucket = '750 to 1000' THEN 4
            WHEN size_bucket = '1000 to 1250' THEN 5
            WHEN size_bucket = '1250 to 1500' THEN 6
            WHEN size_bucket = '1500 to 1750' THEN 7
            WHEN size_bucket = '1750 to 2000' THEN 8
            WHEN size_bucket = '> 2000' THEN 9
        END;
"""
results = pd.read_sql_query(query, engine)
print(results)
```

```python
#Store Age Bucket vs Total Sales
query = """
    SELECT
        CASE
            WHEN strftime('%Y', 'now') - strftime('%Y', open_date) <= 5 THEN '<=5'
            WHEN strftime('%Y', 'now') - strftime('%Y', open_date) BETWEEN 6 AND  10 THEN '5 to 10'
            WHEN strftime('%Y', 'now') - strftime('%Y', open_date) BETWEEN 11 AND 15 THEN '10 to 15'
            WHEN strftime('%Y', 'now') - strftime('%Y', open_date) BETWEEN 16 AND 20 THEN '15 to 20'
        END AS store_age_bucket,
        ROUND(SUM(unit_price_usd * quantity), 2) AS total_sales
    FROM overall
    GROUP BY store_age_bucket
    ORDER BY store_age_bucket
"""
results = pd.read_sql_query(query, engine)
print(results)
```

```python
#Storekey vs Country vs Continent vs State    
query = """
    SELECT
      storekey,Country,Continent,State,
      ROUND(SUM((unit_price_usd)*quantity),2) AS total_revenue_USD
    FROM
      overall
    GROUP BY
      storekey,Country,Continent,State
    ORDER BY total_revenue_USD DESC
"""
results = pd.read_sql_query(query, engine)
print(results)
```

# SALES ANALYSIS

```python
#sales analysis
query = """
    SELECT
      strftime('%m', order_date) AS month,
      ROUND(SUM((unit_price_usd)*quantity),2) AS total_revenue_USD
    FROM
      overall
    GROUP BY
      strftime('%m', order_date)
"""
results = pd.read_sql_query(query, engine)
print(results)
```

```python
#Sales by Top Product Performance
query = """
    SELECT
      product_name,
      SUM(quantity) AS total_quantity
    FROM
       overall
    GROUP BY
      product_name
    ORDER BY total_quantity DESC
    LIMIT 10
"""
results = pd.read_sql_query(query, engine)
print(results)
```

```python
#Sales by Revenue Performance
query = """
    SELECT
      product_name,
      ROUND(SUM((unit_price_usd)*quantity),2) AS total_revenue_USD
    FROM
      overall
    GROUP BY
      product_name
    ORDER BY total_revenue_USD DESC
    LIMIT 10
"""
results = pd.read_sql_query(query, engine)
print(results)
```

```python
#Sales by Stores
query = """
    SELECT
      storekey,
      ROUND(SUM((unit_price_usd)*quantity),2) AS total_revenue_USD
    FROM
      overall
    GROUP BY
      storekey
    ORDER BY total_revenue_USD DESC
"""
results = pd.read_sql_query(query, engine)
print(results)
```

# PRODUCT ANALYSIS

```python
#category and subcategory analysis
query = """
    SELECT
        category, subcategory,
        ROUND(SUM(unit_price_usd * quantity),2) AS total_sales
    FROM overall
    GROUP BY category, subcategory
    ORDER BY total_sales DESC
"""
results = pd.read_sql_query(query, engine)
print(results)
```

```python
#product popularity(top 10 most popular)
query = """
    SELECT
        product_name, SUM(quantity) AS Quantity
    FROM overall
    GROUP BY product_name
    ORDER BY Quantity DESC
    LIMIT 10
"""
results = pd.read_sql_query(query, engine)
print(results)
```

```python
#product popularity(top 10 least popular)
query = """
    SELECT
        product_name, SUM(quantity) AS Quantity
    FROM overall
    GROUP BY product_name
    ORDER BY Quantity ASC
    LIMIT 10
"""
results = pd.read_sql_query(query, engine)
print(results)
```

```python
#profitablity analysis
query = """
    SELECT
        product_name,
        ROUND(SUM((unit_price_usd-unit_cost_usd)*quantity),2) AS Profit_Margin
    FROM overall
    GROUP BY product_name
    ORDER BY Profit_Margin DESC
    LIMIT 10
"""
results = pd.read_sql_query(query, engine)
print(results)
```

# CUSTOMER ANALYSIS 

```python
#Gender count
query = """
    SELECT gender, COUNT(*) AS Count_gender
    FROM customers
    GROUP BY gender
"""
results = pd.read_sql_query(query, engine)
print(results)
```

```python
#Age Bucketing
query = """
    SELECT
        age_bucket,
        COUNT(*) AS count
    FROM (
        SELECT
            *,
            CASE
                WHEN strftime('%Y', order_date) - strftime('%Y', birthday) <= 18 THEN '<=18'
                WHEN strftime('%Y', order_date) - strftime('%Y', birthday) BETWEEN 18 AND 25 THEN '18-25'
                WHEN strftime('%Y', order_date) - strftime('%Y', birthday) BETWEEN 25 AND 35 THEN '25-35'
                WHEN strftime('%Y', order_date) - strftime('%Y', birthday) BETWEEN 35 AND 45 THEN '35-45'
                WHEN strftime('%Y', order_date) - strftime('%Y', birthday) BETWEEN 45 AND 55 THEN '45-55'
                WHEN strftime('%Y', order_date) - strftime('%Y', birthday) BETWEEN 55 AND 65 THEN '55-65'
                ELSE '>65'
            END AS age_bucket
        FROM overall
    ) AS age_groups
    GROUP BY age_bucket
"""
results = pd.read_sql_query(query, engine)
print(results)
```

```python
#country wise customer count
query = """
    SELECT 
        continent, country, state, city, 
        COUNT(CustomerKey) AS customer_count
    FROM 
        CUSTOMERS
    GROUP BY 
        continent, country, state, city
    ORDER BY 
        customer_count DESC
"""
results = pd.read_sql_query(query, engine)
print(results)
```
